# Predicting the Super Bowl LVI Winner

## 3. Data Pre-processing

```python
# Import relevant libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline

import warnings
warnings.filterwarnings(action='ignore')

from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Ridge, ElasticNet
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import TimeSeriesSplit, GridSearchCV
from sklearn.metrics import mean_squared_error
from sklearn.cluster import KMeans
```

```python
# Import data

team_stats = pd.read_csv('../data/processed/SuperBowl_EDA.csv')
team_stats.head(3)
```

Our target is to predict the winner of the Super Bowl game between Los Angeles Rams & Cincinnati Bengals based on the score. The prediction is derived from the listed features of a team. It is important to point out that values of features for both teams on that day is not included. Thus, the prediction model will involve three steps:

**1. Estimate performance statistics of the two teams playing Super Bowl LVI.**

**2. Develop a regression model that accurately determines the team from previous games played in the season.**

**3. Input the estimated feature values in the model to predict the score for each team.**

## 3.1 Estimate teams' performance statistics

The Los Angeles Rams and Cincinnati Bengals did not play against each other during the regular season of every year. Thus, we will have to estimate how much yards and points their offense will gain and how much of yards and points their defense will allow for the years they didn't play against each other. We can do this by analyzing the performance of each of their opponents. E.g, in the case of Rams, we will find the opponent teams that have performance metrics closest to those of the Bengals in their last three games.

This can be done using clustering analysis. We will classify the Rams' opponents into several clusters. Then, we will calculate the average performance of the Bengals in their last three games as 'test data'. This data will be labeled using the cluster model. We will consider the teams that have the same label as 'test' data for further analysis.

```python
team_stats[(team_stats.home == 'Rams') & (team_stats.away == 'Bengals')]
```

```python
# Create list for features

opponent_features = ['away', 'first_downs_away', 'passing_yards_away', 
                     'rushing_yards_away',  'total_yards_away',
                     'rushing_attempts_away', 'fumbles_away', 
                     'int_away', 'turnovers_away','drives_away', 'def_st_td_away',
                     'score_away']

# Select the features for clustering of the Rams' opponents

team_stats_LAR = team_stats[team_stats.home == 'Rams'].loc[:, opponent_features]
team_stats_LAR.head()
```

```python
# Standardize the selected data

scaler = StandardScaler()
LAR_scaled = scaler.fit_transform(team_stats_LAR.drop(columns='away'))
```

```python
# Determine the optimal number of clusters

ssd = []

for c in range(1,len(LAR_scaled)):
    kmeans = KMeans(n_clusters=c, random_state=42)
    kmeans.fit(LAR_scaled)
    ssd.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(30, 7.5))
plt.plot(range(1, len(LAR_scaled)), ssd, marker='o', linestyle='--')
plt.xlabel('# of clusters')
plt.ylabel('sum of squared distances')
plt.xticks(range(0, len(LAR_scaled), 2))
plt.show();
```

'Sum of squared distances' represents distribution of the points in the corresponding cluster. As the number of clusters increas in the figure above, the distribution consistently gets tighter. Thus we will classify the data into 145 clusters. This means only of the Ram's opponents from regular season will ne identified as the closest match to the Bengals' average performance over the last three games.

```python
# Create clusters

kmeans = KMeans(n_clusters=145, random_state=42)
kmeans.fit(LAR_scaled)
```

```python
# Assign cluster labels to Rams' opponents

team_stats_LAR['Label'] = kmeans.labels_
team_stats_LAR
```

```python
# Calculate Bengals' average performance over the last three games
# Use features list without away team feature
opponent_features_without_away = ['first_downs_away', 'passing_yards_away', 
                     'rushing_yards_away',  'total_yards_away',
                     'rushing_attempts_away', 'fumbles_away', 
                     'int_away', 'turnovers_away','drives_away', 'def_st_td_away',
                     'score_away']

CIN_as_Opp = team_stats[team_stats.away == 'Bengals'].loc[:,opponent_features_without_away].iloc[-4:,:]
CIN_as_Opp
```

We need to keep in mind that we are considering the Bengals as the Rams' opponent. Thus, Bengals offensive performance is equivalent to Rams' defensive performance and vice versa. We can flip the above data frame so that the columns are consistent with the Rams' performance statistics.

```python
# Change the column titles

CIN_as_Opp.columns = ['Opp_first_downs', 'Opp_passing_yards', 
                     'Opp_rushing_yards',  'Opp_total_yards',
                     'Opp_rushing_attempts', 'Opp_fumbles', 
                     'Opp_int', 'Opp_turnovers','Opp_drives', 'Opp_def_st_td',
                     'Opp_score']

CIN_as_Opp
```

```python
# Calculate average performance metrics

CIN_as_Opp_stats = CIN_as_Opp.mean(axis=0)
CIN_as_Opp_y = CIN_as_Opp_stats.to_numpy().reshape(1,-1)
CIN_as_Opp_y
```

```python
# Use the clustering model to predict the label of CIN_y data

CIN_as_Opp_label = kmeans.predict(CIN_as_Opp_y)
CIN_as_Opp_label
```

```python
# Find Rams' opponents with label 21

closest_match_LAR = team_stats_LAR[team_stats_LAR.Label == 21]
closest_match_LAR
```

**The Bengals performance in the last three games is closest to the Chiefs performance against the Rams**

```python
# Add the closest match to recent three performances of the Bengals

CIN_as_Opp = CIN_as_Opp.append(closest_match_LAR.drop(columns=['away', 'Label']))
CIN_as_Opp
```

```python
# We will input the average of CIN_as_opp values in the regression model to predict score

LAR_performance = CIN_as_Opp.mean(axis=0)
LAR_performance
```

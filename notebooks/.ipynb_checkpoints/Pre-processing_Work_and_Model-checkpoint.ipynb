# Predicting the Super Bowl LVI Winner

## 3. Data Pre-processing

```python
# Import relevant libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline

import warnings
warnings.filterwarnings(action='ignore')

from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Ridge, ElasticNet
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.model_selection import TimeSeriesSplit, GridSearchCV
from sklearn.metrics import mean_squared_error
from sklearn.cluster import KMeans
```

```python
# Import data

team_stats = pd.read_csv('../data/processed/SuperBowl_EDA.csv')
team_stats.head(3)
```

Our target is to predict the winner of the Super Bowl game between Los Angeles Rams & Cincinnati Bengals based on the score. The prediction is derived from the listed features of a team. It is important to point out that values of features for both teams on that day is not included. Thus, the prediction model will involve three steps:

**1. Estimate performance statistics of the two teams playing Super Bowl LVI.**

**2. Develop a regression model that accurately determines the team from previous games played in the season.**

**3. Input the estimated feature values in the model to predict the score for each team.**

### 3.1 Estimate teams' performance statistics

The Los Angeles Rams and Cincinnati Bengals did not play against each other during the regular season of every year. Thus, we will have to estimate how much yards and points their offense will gain and how much of yards and points their defense will allow for the years they didn't play against each other. We can do this by analyzing the performance of each of their opponents. E.g, in the case of Rams, we will find the opponent teams that have performance metrics closest to those of the Bengals in their last three games.

This can be done using clustering analysis. We will classify the Rams' opponents into several clusters. Then, we will calculate the average performance of the Bengals in their last three games as 'test data'. This data will be labeled using the cluster model. We will consider the teams that have the same label as 'test' data for further analysis.

```python
team_stats[(team_stats.home == 'Rams') & (team_stats.away == 'Bengals')]
```

```python
# Create list for features

opponent_features = ['away', 'first_downs_away', 'passing_yards_away', 
                     'rushing_yards_away',  'total_yards_away',
                     'rushing_attempts_away', 'fumbles_away', 
                     'int_away', 'turnovers_away','drives_away', 'def_st_td_away',
                     'score_away']

# Select the features for clustering of the Rams' opponents

team_stats_LAR = team_stats[team_stats.home == 'Rams'].loc[:, opponent_features]
team_stats_LAR.head()
```

```python
# Standardize the selected data

scaler = StandardScaler()
LAR_scaled = scaler.fit_transform(team_stats_LAR.drop(columns='away'))
```

```python
# Determine the optimal number of clusters

ssd = []

for c in range(1,len(LAR_scaled)):
    kmeans = KMeans(n_clusters=c, random_state=42)
    kmeans.fit(LAR_scaled)
    ssd.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(30, 7.5))
plt.plot(range(1, len(LAR_scaled)), ssd, marker='o', linestyle='--')
plt.xlabel('# of clusters')
plt.ylabel('sum of squared distances')
plt.xticks(range(0, len(LAR_scaled), 2))
plt.show();
```

'Sum of squared distances' represents distribution of the points in the corresponding cluster. As the number of clusters increas in the figure above, the distribution consistently gets tighter. Thus we will classify the data into 145 clusters. This means only of the Ram's opponents from regular season will ne identified as the closest match to the Bengals' average performance over the last three games.

```python
# Create clusters

kmeans = KMeans(n_clusters=145, random_state=42)
kmeans.fit(LAR_scaled)
```

```python
# Assign cluster labels to Rams' opponents

team_stats_LAR['Label'] = kmeans.labels_
team_stats_LAR
```

```python
# Calculate Bengals' average performance over the last three games
# Use features list without away team feature
opponent_features_without_away = ['first_downs_away', 'passing_yards_away', 
                     'rushing_yards_away',  'total_yards_away',
                     'rushing_attempts_away', 'fumbles_away', 
                     'int_away', 'turnovers_away','drives_away', 'def_st_td_away',
                     'score_away']

CIN_as_Opp = team_stats[team_stats.away == 'Bengals'].loc[:,opponent_features_without_away].iloc[-4:,:]
CIN_as_Opp
```

We need to keep in mind that we are considering the Bengals as the Rams' opponent. Thus, Bengals offensive performance is equivalent to Rams' defensive performance and vice versa. We can flip the above data frame so that the columns are consistent with the Rams' performance statistics.

```python
# Change the column titles

CIN_as_Opp.columns = ['Opp_first_downs', 'Opp_passing_yards', 
                     'Opp_rushing_yards',  'Opp_total_yards',
                     'Opp_rushing_attempts', 'Opp_fumbles', 
                     'Opp_int', 'Opp_turnovers','Opp_drives', 'Opp_def_st_td',
                     'Opp_score']

CIN_as_Opp
```

```python
# Calculate average performance metrics

CIN_as_Opp_stats = CIN_as_Opp.mean(axis=0)
CIN_as_Opp_y = CIN_as_Opp_stats.to_numpy().reshape(1,-1)
CIN_as_Opp_y
```

```python
# Use the clustering model to predict the label of CIN_y data

CIN_as_Opp_label = kmeans.predict(CIN_as_Opp_y)
CIN_as_Opp_label
```

```python
# Find Rams' opponents with label 21

closest_match_LAR = team_stats_LAR[team_stats_LAR.Label == 21]
closest_match_LAR
```

**The Bengals performance in the last three games is closest to the Chiefs performance against the Rams**

```python
# Add the closest match to recent three performances of the Bengals

CIN_as_Opp = CIN_as_Opp.append(closest_match_LAR.drop(columns=['away', 'Label']))
CIN_as_Opp
```

```python
# We will input the average of CIN_as_opp values in the regression model to predict score

LAR_performance = CIN_as_Opp.mean(axis=0)
LAR_performance
```

### 3.2 Develop a regression model to predict scores

#### 3.2.1 Data pre-processing

```python
# Inspect data frame 

team_stats.head()
```

```python
# Inspect column names

team_stats.columns
```

```python
# Inspect data types

team_stats.info()
```

```python
# Convert date into a datetime object

team_stats['date'] = pd.to_datetime(team_stats['date'])
team_stats.info()
```

```python
# Drop 'away' column as we already have the 'home' column for regression analysis
# Create columns to drop including 'away'

columns_to_drop = ['away', 'third_downs_away', 'third_downs_home',
                   'fourth_downs_away', 'fourth_downs_home',
                   'redzone_away', 'redzone_home',
                   'possession_away', 'possession_home',
                   'penalties_away', 'penalties_home',
                   'comp_att_away', 'comp_att_home',
                   'sacks_away','sacks_home']
                   
team_stats_reg = team_stats.drop(columns=columns_to_drop)
```

```python
# Create list for features compatible with regression modedeling

team_stats_reg.head(5)
```

```python
# Create dummy variables for team names

team_names = pd.get_dummies(team_stats_reg['home'], drop_first=True)
team_names.head(3)
```

```python
# Remove 'home' feature from the data frame and append dummy variables instead

team_stats_reg = team_stats_reg.drop(columns='home')
team_stats_reg = pd.concat([team_stats_reg, team_names], axis=1)
team_stats_reg.head(3)
```

```python
# Sort the data by 'date'

team_stats_reg = team_stats_reg.sort_values(by='date', axis=0)
team_stats_reg
```

```python
# Split features and target variables

X = team_stats_reg.drop(columns=['score_away', 'score_home'])
y = team_stats_reg['score_home']
```

```python
# Split training and test datasets

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=30, random_state=42)
```

```python
# Drop date column

X_train.drop(columns='date', inplace=True)
X_train.columns
```

```python
# Standardize numerical inputs

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_train)
X_scaled
```

#### 3.2.2. Linear regression model

```python
# Invistigate shape of data frame

X_train.shape
```

Since the data frame contains 51 features and 5327 entries, we will use 'Ridge' and 'ElasticNet' regressors that punish large weights of the features along with Random Forest Regressor. We intend to predict the next week's data using previous weeks' entries.

```python
# Instantiate 'Ridge' and Predict 'y'

ridge = Ridge()
ridge.fit(X_train, y_train)
y_pred = ridge.predict(X_train)
print(f'mean squared error: {mean_squared_error(y_train, y_pred)}')
```

```python
# Plot the predictions

plt.figure(figsize=(15,7.5))
plt.scatter(y_train, y_pred)
plt.plot(range(60), range(60), linewidth=2, color='red')
plt.xlabel('y')
plt.ylabel('y_pred')
plt.show()
```

Model training and cross-validation:

Since the game statistics are generated over many weeks, this data must be considered time-dependent. Thus we need to make sure the training set consist of the older data and validation set consist of the recent data while cross-validating the models. This can be achieved by 'TimeSeriesSplit' as the cross-validator. "TimeSeriesSplit' is similar to 'KFold' but uses first 'K' folds as the training set and (k+1)th fold as the validation set.

Choosing the number of splits is also critical. We want to predict a team's performance in the current week based on its performance in the weeks prior. Not every team plays a game every week during the regular season also known as a "bye week". Thus, if we choose slices of length 32 (total number of teams in the NFL), some slices will feature one or more teams twice (slices span across two weeks). On the other hand, if we choose much smaller slices (e.g., 15 entries), we end up using the performance of some teams to predict the performance of other teams that played in the same week (Kth and (k+1)th folds are in the same week).

The best way to compromise is to select the slice length as the minimum number of teams that played during the regular season, i.e. 26 of each season. Since 5327 (total entries)/ 568 (average entries per year) = 9.38, we will multiply that times 568 / 26 (which is 21.85). The data will be split into 204.95 slices during cross validation.

```python
# Use GridSearch to optimize 'Ridge' hyperparameters

param_grid = {'alpha':[0.005, 0.01, 0.05, 0.1, 0.5, 1]}

# Round the n_splits to an integer (204.95 to 205)

tscv = TimeSeriesSplit(n_splits=205)

ridge_grid = GridSearchCV(estimator=ridge, param_grid=param_grid, cv=tscv)

ridge_grid.fit(X_train, y_train)
```

```python
# Print optimal hyperparameter value

print(f'Best hyperparameter value for the Ridge model is"{ridge_grid.best_params_}')
```

```python
# Predict 'y' using cross-validated data

y_pred_ridge = ridge_grid.predict(X_train)
print(f'mean squared error:{mean_squared_error(y_train, y_pred_ridge)}')
```

```python
# Plot the predictions

plt.figure(figsize=(15,7.5))
plt.scatter(y_train, y_pred_ridge)
plt.plot(range(60),range(60), linewidth=2, color='red')
plt.xlabel('y')
plt.ylabel('y_pred_ridge')
plt.show()
```

```python
# Use GridSearch to optimize the 'ElasticNet' hyperparameters

enet = ElasticNet()

param_grid = {'alpha':[0.05, 0.1, 0.5, 1], 'l1_ratio':[0.05, 0.1, 0.5, 1]}
tscv = TimeSeriesSplit(n_splits=205)

enet_grid = GridSearchCV(estimator=enet, param_grid=param_grid, cv=tscv)

enet_grid.fit(X_train, y_train)
```

```python
# Print optimal hyperparameter values

print(f'Best hyperparameter values for the ElasticNet model is: {enet_grid.best_params_}')

# Predict 'y' using cross_validated data

y_pred_enet = enet_grid.predict(X_train)
print(f'mean squared error:{mean_squared_error(y_train, y_pred_enet)}')
```

```python
# Plot the predictions

plt.figure(figsize=(15,7.5))
plt.scatter(y_train, y_pred_enet)
plt.plot(range(60),range(60), linewidth=2, color='red')
plt.xlabel('y')
plt.ylabel('y_pred_enet')
plt.show()
```

```python
# Use GridSearch to optimize the 'RandomForestRegressor' hyperparameters

rf = RandomForestRegressor()

param_grid = {'n_estimators':[100, 200, 300, 400, 500]}
tscv = TimeSeriesSplit(n_splits=205)

rf_grid = GridSearchCV(estimator=rf, param_grid=param_grid, cv=tscv)

rf_grid.fit(X_train, y_train)
```

```python
# Print optimal hyperparameter values

print(f'Best hyperparameter values for the Random Forest Regressor model is: {rf_grid.best_params_}')

# Predict 'y' using cross_validated data

y_pred_rf = rf_grid.predict(X_train)
print(f'mean squared error:{mean_squared_error(y_train, y_pred_rf)}')
```

```python
# Plot the predictions

plt.figure(figsize=(15,7.5))
plt.scatter(y_train, y_pred_rf)
plt.plot(range(60),range(60), linewidth=2, color='red')
plt.xlabel('y')
plt.ylabel('y_pred_rf')
plt.show()
```
